```python
df_uptime['uptime'] = pd.to_numeric(df_uptime['uptime'])
df_uptime['stake'] = pd.to_numeric(df_uptime['stake'])
```

```python
# df_uptime = df_uptime[(df_uptime != 0).all(axis=1)]
# print(df_uptime)
Stake_weighted_uptime_o    = sum(df_uptime['uptime'] * df_uptime['stake'])/  sum( df_uptime['stake'] )
print(Stake_weighted_uptime_o)
simple_uptime_o = df_uptime['uptime'].mean()
print(simple_uptime_o)
```

```python
df_validators.Credits_earned              = df_validators.Credits_earned.astype(str).astype(float)
df_validators['staked amount']            = df_validators['staked amount (lamp)'].astype(str).astype(float)

df_validators.drop(df_validators.columns.difference(['Credits_earned','points_earned (sol)','staked amount','epoch tot rewards']), 1, inplace=True)

try:
  df_epoch.rename(index = lambda s:df_epoch.Parameter[s],inplace=True)
except:
  print('Already run this code once')

df_validators['Expected Rewards']               = df_validators.apply(lambda _: '', axis=1)
df_validators['Expected APY']                   = df_validators.apply(lambda _: '', axis=1)
df_validators['Absolute Difference in Rewards'] = df_validators.apply(lambda _: '', axis=1)
df_validators['% Difference in Rewards']        = df_validators.apply(lambda _: '', axis=1)
df_validators['uptime']                         = df_validators.apply(lambda _: '', axis=1)

df_epoch.Value.slotsPerEpoch  = float(df_epoch.Value.slotsPerEpoch)
df_validators['uptime']       = df_validators.Credits_earned  / df_epoch.Value.slotsPerEpoch


df_epoch.drop(df_epoch.columns.difference(['Value']), 1, inplace=True)

df_validators                                = df_validators.reindex(columns = ['Credits_earned','points_earned (sol)','staked amount', 'uptime','epoch tot rewards', 'Expected Rewards', 'Expected APY', 'Absolute Difference in Rewards', '% Difference in Rewards'])
```

```python
# Constants from Network
uptime_o            = Stake_weighted_uptime_o
#uptime_o            = float(df_epoch.Value['uptime^o_t'])
total_supply        = float(df_epoch.Value['total_supply_t'])/1E9
SOL_staked_Pct      = float(df_epoch.Value['%SOL_staked'])
slot_time           = float(df_epoch.Value['slot_time'])*1000  
EPOCHS_PER_YEAR       = float(df_epoch.Value['EPOCHS_PER_YEAR'])
TARGET_TIME           = float(df_epoch.Value['TARGET_TIME'])    # ms
SPS_TARGET            = float(df_epoch.Value['SPS_TARGET'])
LAMPORTS_PER_SIGNATURE  = float(df_epoch.Value['LAMPORTS_PER_SIGNATURE'])      # lamports
slotsPerEpoch         = float(df_epoch.Value['slotsPerEpoch']) 
FEE_BURN_PCT          = float(df_epoch.Value['FEE_BURN_PCT'])/100 
INITIAL_INFLATION_RATE  = float(df_epoch.Value['INITIAL_INFLATION_RATE']) 
DISINFLATION_RATE       = float(df_epoch.Value['DISINFLATION_RATE'] )
LONG_TERM_INFLATION_RATE= float(df_epoch.Value['LONG_TERM_INFLATION_RATE'] )
INFLATION_RATE        = float(df_epoch.Value['INFLATION_RATE'] )
sps_t                 = float(df_epoch.Value['sps_t']) 
```

```python
# uptime_o = simple_uptime_o
```

```python
# Parameters to be varied by validator
for i in range(len(df_validators)):
# for i in range(1):
  stake               = float(df_validators['staked amount'][i])/1E9
  uptime_v              = float(df_validators['uptime'][i])

  # Calculations of the current epoch
  # E_validator_rewards =  ( (1+INFLATION_RATE)**(1/EPOCHS_PER_YEAR)-1 )  * total_supply/SOL_staked_Pct 
  E_validator_rewards =  INFLATION_RATE/EPOCHS_PER_YEAR  * total_supply 
  E_validator         = stake / (SOL_staked_Pct*total_supply) * uptime_v /uptime_o* E_validator_rewards   # IGNORED RENT
  E_txFees            = slotsPerEpoch *(1- FEE_BURN_PCT) * LAMPORTS_PER_SIGNATURE * 1E-09 * pow(sps_t,2) / SPS_TARGET
  E_MEV               = 0
  E_slotLeader        = E_validator + E_txFees + E_MEV

  P_slotLeader = stake/(SOL_staked_Pct*total_supply)  
  P_validator = (1-P_slotLeader)
  P_slotLeader= P_slotLeader

  E_rewards = P_slotLeader*E_slotLeader + P_validator*E_validator
  df_validators['Expected Rewards'][i] = E_rewards
  df_validators['Absolute Difference in Rewards'][i] = E_rewards - float(df_validators['epoch tot rewards'][i])
  df_validators['% Difference in Rewards'][i] = (E_rewards - float(df_validators['epoch tot rewards'][i])) / float(df_validators['epoch tot rewards'][i]) *100

  # print("P_slotLeader:", P_slotLeader)
  # print("E_slotLeader:", E_slotLeader)
  # print("The expected slotLeader reward given to a validator over the epoch is: "+ str(P_slotLeader*E_slotLeader) )
  # print("The expected validator rewards given to a validator over the epoch is: "+ str(P_validator*E_validator) )
  # print("The expected total rewards given to a validator over the epoch is: " + str(E_rewards) )
  APY = E_rewards/stake*EPOCHS_PER_YEAR*TARGET_TIME/slot_time
  df_validators['Expected APY'][i] = APY
  # print("The expected APY for the year is: "+ str(APY) )
  # print(i)
```
